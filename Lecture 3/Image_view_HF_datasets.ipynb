```python
import matplotlib.pyplot as plt
from datasets import load_dataset
from collections import defaultdict


def example_usage():
    # Load the dataset
    tiny_imagenet = load_dataset('Maysee/tiny-imagenet', split='train')

    # Group images by label
    images_by_class = defaultdict(list)
    for item in tiny_imagenet:
        label = item['label']
        if len(images_by_class[label]) < 5:
            images_by_class[label].append(item['image'])

    # Display 5 images per class for the first 5 classes
    for label, images in list(images_by_class.items())[:5]:
        plt.figure(figsize=(10, 2))
        plt.suptitle(f"Class Label: {label}", fontsize=16)
        for i, img in enumerate(images):
            plt.subplot(1, 5, i + 1)
            plt.imshow(img)
            plt.axis('off')
        plt.show()

if __name__ == '__main__':
    example_usage()
```
